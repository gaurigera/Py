# One Hot Encoding

```python
num_cols=data.select_dtypes(include='number').columns

cat_cols=data.columns.difference(num_cols)
cat_cols=cat_cols[1:]
```

```python
cat_cols #All the columns without any numeric data
```

```python
##Creating a dataframe with only the categorical data
data_cat=data[cat_cols].copy()
data_cat.head()
```

```python
data_cat = pd.get_dummies(data_cat,columns=cat_cols)
```

```python
data_cat.head()
```

```python
data=pd.concat([data,data_cat],axis=1)
data.head()
```

```python
data.drop(columns=cat_cols,inplace=True)
```

# Data Normalization

```python
num_col=list(data.select_dtypes(include="number").columns)
num_col.remove('id')
num_col.remove('label')
```

```python
from sklearn.preprocessing import MinMaxScaler
# using minmax scaler for normalizing data
minmax_scale = MinMaxScaler()
def normalization(df,col):
  for i in col:
    arr = df[i]
    arr = np.array(arr)
    df[i] = minmax_scale.fit_transform(arr.reshape(len(arr),1))
  return df
```

```python
normalization(data,num_col)
data.head()
```

# Label Encoding

**Binary Labels**

```python
#Changing attack labels (normal and abnormal instead of 0 and 1)
bin_label = pd.DataFrame(data.label.map(lambda x: 'normal' if x==0 else 'abnormal'))
```

```python
bin_data= data.copy()
bin_data['label']=bin_label
```

```python
from sklearn import preprocessing


le1= preprocessing.LabelEncoder()
encode_label=bin_label.apply(le1.fit_transform)
bin_data['label']=encode_label
```

```python
le1.classes_ #Distinct classes encountered
```

```python
np.save("le1_classes.npy", le1.classes_ ,allow_pickle=True)
```

# Multi-Class Labels

```python
multi_label=pd.DataFrame(data.attack_cat)
multi_data=data.copy()
```

```python
encode_label=multi_label.apply(le1.fit_transform)
multi_data['label']=encode_label
```
